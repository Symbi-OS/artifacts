```python
import pandas as pd
import seaborn as sb

sb.set_theme(style="darkgrid")
sb.color_palette("Set2")
sb.set(rc={'figure.figsize':(12.7, 7.27)})

def preprocess_data(data: pd.DataFrame):
    base_tp_val = 0
    for i in range(len(data['redis_instances'])):
        if i == 0:
            base_tp_val = float(data['throughput'][i])
            data['throughput'][i] /= base_tp_val
            
df = pd.read_csv("redis_results.csv")
#preprocess_data(df)

g = sb.barplot(data=df, x='type', y='throughput', width=0.5)
g.set(title='Comparing Redis Throughputs')
g.legend(loc='upper right')
g.set_ylabel('Throughput (requests)')
g.set_xlabel('System')

for i in range(len(g.containers)):
    g.bar_label(g.containers[i])
```
